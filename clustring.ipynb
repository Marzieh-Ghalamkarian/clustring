```python
xxx=data[['Gender','Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
from sklearn.preprocessing import MinMaxScaler
xo=xxx.values
scaler=MinMaxScaler()
normaldata=scaler.fit_transform(xo)
normaldata[0:5]
```

```python
from sklearn.metrics.pairwise import euclidean_distances
dis=euclidean_distances(normaldata, normaldata)
print(dis[:5])
```

```python
from sklearn.cluster import AgglomerativeClustering

agglom=AgglomerativeClustering(n_clusters=6,linkage='complete')
rr=agglom.fit(dis)
```

```python
agglom.labels_
```

```python
data['cluster']=agglom.labels_
```
